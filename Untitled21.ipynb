```python
import pandas as pd
import numpy as np

# Creat dictionary to hold data
exam_data = {'Name':['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew', 'Laura', 'Kevin', 'Jonas'],
             'Score':[12.5, 9, 16.5, np.NAN, 9, 20, 14.5, np.NAN, 8, 19],
             'Attempts':[1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
             'Qualify':['Yes', 'No', 'Yes', 'No', 'No', 'Yes', "Yes", 'No', 'No', 'yes'],
             }

# Create labels list to be the index of the DataFrame
labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']

# Create data fram from the 'exam_data' dictionary
# The indices are the labels list
df = pd.DataFrame(exam_data,index=labels)

# Print the first three rows
print(df.head(3))
print()

# Delete rows with NaN values
df = df.dropna()

# Extract the 'Name' and 'Score' Column from the DataFrame
print(df['Name'])
print()
print(df['Score'])
print()

dict2 = {'Name':['Suresh'],
         'Score':[15.5],
         'Attempts':[1],
         'Qualify':['Yes']}

# Append new row 'k' to the DataFrame
df.loc['k'] = ['Suresh', 15.5, 1, 'Yes']
print(df)
print()

# Delete the 'Attempts' column
df = df.drop(columns='Attempts')
print(df)
print()

# Add new column 'Success'
df['Success'] = np.where(df['Score'] > 10, 1, 0)
print(df)
print()

# Export the data frame into a CSV
df.to_csv("my_data.csv")


```
